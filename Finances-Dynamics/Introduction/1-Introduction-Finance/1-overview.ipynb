## 1. The Financial Perspective: Prospective Analysis

### Formal Definition
In Finance, the value of a firm or asset is derived exclusively from a **prospective analysis**. The value of an asset does not correspond to the historical cost invested in it (sunk cost), but rather to the present value of the benefits (cash flows) it will generate in the future.

### In Other Words
Accounting looks in the rearview mirror (what did we spend?), while Finance looks through the windshield (what will we earn?). If you spent $1 million building a factory that generates $0, the financial value is $0, regardless of the investment. Money already spent is a "sunk cost" and is irrelevant for future decision-making.

### Formulas / Syntax / Key Structures
* **Concept:** $Value_0 = f(Future\ Benefits)$
* **Rule:** Ignore Sunk Costs. HOWEVER, initial investments affect future cash flows.
* **Focus:** Sustainability is important (stakeholders, environment) because it ensures long-term cash flow generation, not just for ethical reasons.

### Intuition & Development
The firm is viewed as a set of contracts and assets designed to generate wealth. To value it, we must estimate what comes *next*, not what happened *before*.

## 2. Cash Flows vs. Accounting Profits

### Formal Definition
**Cash Flow (FC)** represents the actual inflow and outflow of cash in a business. <br>
**Accounting Profit (Utility)** is a measure of performance based on accrual accounting principles (recognizing revenue when earned, not when received).

### In Other Words
"Profit" is an opinion; Cash is a fact. You can have high accounting profits but go bankrupt if you have no cash to pay your bills. Conversely, you can have negative profits (due to high depreciation) but positive cash flow.

### Formulas / Syntax
The basic bridge between the two:
$$ Cash\ Flow \approx Net\ Income + Non\_Cash\_Expenses (Depreciation) - \Delta Working\ Capital - CapEx $$

### Intuition
Finance centers on Cash Flow because that is what investors actually receive and can reinvest.

### Example Profit vs. Cash Flow

- Scenario: We sell 5 products for $100 each. 
- 3 customers pay immediately (Cash).
- 2 customers pay on credit (next year).
- Cost per product was $50 (paid in cash).

$$Revenue = 5 * \$100 = \$500$$
$$Costs = 5 * \$50 = \$250$$
$$Accounting\ Profit = Revenue - Costs = \$500 - \$250 = \$250$$
$$Cash\ Flow = Cash\ Received - Cash\ Paid = (3 * \$100) - (5 * \$50) = \$300 - \$250 = \$50$$

```python
# DO NOT MODIFY THIS SNIPPET

revenue = 5 * 100
costs = 5 * 50

accounting_profit = revenue - costs

# Cash Flow Calculation
cash_inflow = 3 * 100  # Only 3 paid
cash_outflow = 5 * 50  # We paid full costs
net_cash_flow = cash_inflow - cash_outflow

print(f"Accounting Profit: ${accounting_profit}")
print(f"Net Cash Flow:     ${net_cash_flow}")

# Notice: We have a Profit of $250, but only $50 cash in the bank.
```

## 3. Risk: Definition and Measurement

### Formal Definition
Risk is the probability that *actual returns* will deviate from *expected returns*. It encompasses both "upside" (doing better than expected) and "downside" (doing worse). Human beings exhibit **Risk Aversion**, meaning they require a higher expected return to accept higher uncertainty.

### In Other Words
Risk isn't just "losing money." It's the rollercoaster ride. If you expect a 10% return, getting 30% is a "risk" outcome just as getting -10% is. Because we dislike uncertainty, we charge a "premium" (higher price/rate) to take it on.

### Key Structures
* **Step 1:** Define risk (Good and Bad outcomes).
* **Step 2:** Determine whose perspective matters (The diversified investor).
* **Step 3:** Translate risk into a number (usually a discount rate).

### Intuition
There is no return without risk. They are correlated. If an investment offers high returns with "zero risk," it is likely a scam or an arbitrage opportunity that will disappear quickly.

## 4. The Time Value of Money (TVM)

### Formal Definition
The principle that a unit of currency received today is worth more than the same unit received in the future due to its potential earning capacity. This is driven by time preference, inflation, and risk.

### In Other Words
The "Mantra" of Finance: **A dollar today is worth more than a dollar tomorrow.** 
Why? 
1. You can spend it now (Preference).
2. Prices go up (Inflation).
3. You might not get paid tomorrow (Risk).

### Formulas / Syntax

---

**Present Value Formula: (PV)**
$$ PV = \frac{FV}{(1 + r)^n} $$

---

Where:
* $PV$: Present Value (Value today)
* $FV$: Future Value (Cash flow in the future)
* $r$: Discount rate (reflecting risk and opportunity cost)
* $n$: Number of periods

### Intuition
This is fundamentally about **Opportunity Cost**. If I lend you money, I lose the opportunity to invest it elsewhere. You must pay me for that lost opportunity.

### Example:

Lets calculate the Present Value of a future cash flow. <br>
FV: Amount to be received in the future ==> fv = 1000 <br>
r: Discount rate (e.g., 0.10 for 10%)   ==> r = 0.10 <br>
n: Number of years                      ==> n = 5 <br>

**Remember the formula:**
$$ PV = \frac{FV}{(1 + r)^n} = \frac{1000}{(1 + 0.10)^5} = \frac{1000}{(1.1)^5} = \frac{1000}{1.61051} â‰ˆ 620.92 $$

```python
# DO NO MODIFY THIS SNIPPET
# Implementation of Time Value of Money

def calculate_present_value(future_value, rate, years):
    """
    Calculates the Present Value of a future cash flow.
    FV: Amount to be received in the future
    r: Discount rate (e.g., 0.10 for 10%)
    n: Number of years
    """
    return future_value / ((1 + rate) ** years)

# Example: receiving $1,000 in 5 years with a 10% discount rate
fv = 1000
r = 0.10
n = 5

pv = calculate_present_value(fv, r, n)
print(f"Value of ${fv} received in {n} years at {r*100}% rate: ${pv:.2f}")

print("In the following graph, we visualize how the present value of $1000 decreases as the number of years increases, given a discount rate of 10%.")

# Visualization of decaying value over time
years_range = np.arange(0, 21)
pv_values = [calculate_present_value(1000, 0.10, y) for y in years_range]

plt.figure(figsize=(10, 5))
plt.plot(years_range, pv_values, marker='o')
plt.title('Present Value of $1000 received in year n (Rate=10%)')
plt.xlabel('Years (n)')
plt.ylabel('Present Value ($)')
plt.show()
```

## 5. Asset Valuation & Financial Markets

### Formal Definition
**Valuation:** The value of *any* asset is the Present Value of its expected future cash flows.<br>
**Financial Markets:** The mechanism that facilitates the raising of capital and the transfer of risk. They enforce the "Law of One Price" through **Arbitrage**.

### In Other Words
* **Valuation:** To know what a stock is worth, you sum up all the discounted money it will pay you in the future.
* **Markets:** There are no free lunches. If there is easy money to be made (arbitrage), everyone will jump in, prices will adjust, and the opportunity will disappear. We assume markets are competitive but have frictions (taxes, transaction costs).

### Formulas / Syntax
**General Valuation Equation:**
$$ V_0 = \sum_{t=1}^{T} \frac{E(CF_t)}{(1+r)^t} $$

### Intuition
Different assets have different cash flow structures:
1.  **Debt:** Contractual flows (Interest). Discount rate = Default risk.
2.  **Equity:** Residual flows (Dividends/Buybacks). Discount rate = Perceived market risk.
3.  **Options:** Contingent flows (Depends on another asset).

# Example: Valuing a simple asset (a bond)

Asset pays $100 every year for 3 years, then pays back $1000 principal. <br>
Discount rate is 5%.

```python
# DO NOT MODIFY THIS SNIPPET

cash_flows = [100, 100, 1100] # Year 1, Year 2, Year 3 (Coupon + Principal)
rate = 0.05

value = 0
for t, cf in enumerate(cash_flows, start=1):
    pv_cf = cf / ((1 + rate) ** t)
    value += pv_cf
    print(f"Year {t} CF: ${cf} -> PV: ${pv_cf:.2f}")

print(f"\nTotal Asset Value: ${value:.2f}")
```
