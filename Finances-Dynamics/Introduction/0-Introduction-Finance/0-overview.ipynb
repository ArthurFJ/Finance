# Introduction to the Corporate Financial World

## The core Philosophy of Finance

In business, while qualitative factors like environment, or social impact are important for long term sustainability, the ultimate measure of success is the generation of value through money.

### Definition 

**Formal**: Finance is the discipline concerned with the valuation of assets based on their capacity to generate future cash flows. Finance is prospective. Unlike accounting, which is retrospective and focuses on historical costs and revenues, finance looks forward to estimate the present value of future cash flows.

**In other words**: It's all about the money! It all comes down to how much cash a business or transaction can generate as income or cash flow in the future.

## The Pillars of Finance

I. **Prospective Analysis**

- From the assets perspective, a business is valued based on the future potential to generate cash flows, not on past performance like in accounting. 
- Money is already spent, so it is a "sunk cost". This means, the value of an asset is determined by the expected future cash flow it can make.
